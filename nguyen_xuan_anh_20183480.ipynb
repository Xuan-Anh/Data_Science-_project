````python
import wordcloud
import re
````

````python
generate_wordcloud = wordcloud.WordCloud()
for month in range(1,13):
    words = " ".join(df[df['release_month'] == month]['title'])
    words = words.replace("game", " ")
    words = words.replace("Game", " ")
    words = words.replace("edition", " ")
    words = words.replace("Edition", " ")
    # words = " ".join(words)
    word_cloud = generate_wordcloud.generate(words)
    plt.imshow(word_cloud)
    plt.title('Month ' + str(month))
    plt.show()
    # break
````

How are the releases by top genres by years

````python
year_list = df['release_year'].unique().tolist()
print(year_list )
type(year_list[2])
year_list = [year for year in year_list if year > 0  ]
year_list.sort()
````

````python
top_genre_by_year = {}
for year in year_list:
    top = df[df['release_year'] == year]['genre'].value_counts()[0:1]
    top_genre_by_year[year] = [top.index[0] , top.values[0]]
# print(top_genre_by_year)
year_name = []
count = []
for year in top_genre_by_year:
    year_name.append(str(int(year))+ "_" + top_genre_by_year[year][0])
    count.append(top_genre_by_year[year][1])
# plt.fi
plt.figure(figsize = (15,10))
sns.barplot(x = year_name, y = count)
plt.xticks(rotation=70)
plt.show()

````

==> Action was the most genre in almost year

Platforms with highest releases by year

````python
top_genre_by_year = {}
for year in year_list:
    top = df[df['release_year'] == year]['platform'].value_counts()[0:1]
    top_genre_by_year[year] = [top.index[0] , top.values[0]]
# print(top_genre_by_year)
year_name = []
count = []
for year in top_genre_by_year:
    year_name.append(str(int(year))+ "_" + top_genre_by_year[year][0])
    count.append(top_genre_by_year[year][1])
# plt.fi
plt.figure(figsize = (15,10))
sns.barplot(x = year_name, y = count)
plt.xticks(rotation=70)
plt.show()
````
